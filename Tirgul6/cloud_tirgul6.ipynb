```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import db

# initialize Firebase credentials
cred = credentials.Certificate('/content/drive/MyDrive/Colab Notebooks/CloudCourse/Tirgul6/cloudtirgul6-firebase-adminsdk-3m2sb-5d58089424.json')



# check if a default app already exists
if not firebase_admin._apps:
    # if a default app doesn't exist, initialize one with a unique name
    firebase_admin.initialize_app(cred, {
        'databaseURL': 'https://cloudtirgul6-default-rtdb.firebaseio.com/'
    }, name='Tirgul6')
    
# create a reference to the questions node
ref = db.reference('questions')
questions = ref.get()
```

```python
question_nodes = ref.get()

# create empty lists to store the questions, answers, and correct answers
question_list = []
answer_list = []
correct_answer_list = []

# loop through the question nodes and extract the question, answers, and correct answer for each question
for question_id, question_data in question_nodes.items():
    question_list.append(question_data['question'])
    answer_list.append(question_data['answers'])
    correct_answer_list.append(question_data['correct_answer'])

# initialize the score and index of the current question
score = 0
current_question_index = 0

# loop through all the questions
while current_question_index < len(question_list):
    
    # get the current question and answers
    current_question = question_list[current_question_index]
    current_answers = answer_list[current_question_index]
    
    # display the current question and answers to the user
    print(current_question)
    for i, answer in enumerate(current_answers):
        print(f"{i+1}. {answer}")
        
    # get the user's answer
    user_answer = input("Enter the number of your answer (1-4): ")
    
    # check if the user's answer is correct and update the score accordingly
    if current_answers[int(user_answer)-1] == correct_answer_list[current_question_index]:
        print("Correct!")
        score += 10
    else:
        print("Wrong.")
        
    # move on to the next question
    current_question_index += 1
    
    # check if there are no more questions left and end the game if necessary
    if current_question_index == len(question_list):
        print("End of game.")
    else:
        play_again = input("Press Enter to continue to the next question, or type 'end' to quit the game: ")
        if play_again == "end":
            print("End of game.")
            break
```

```python
# get the questions data from Firebase and extract the questions, answers, and correct answers
question_nodes = ref.get()
question_list = []
answer_list = []
correct_answer_list = []

for question_id, question_data in question_nodes.items():
    question_list.append(question_data['question'])
    answer_list.append(question_data['answers'])
    correct_answer_list.append(question_data['correct_answer'])

def check_answer(user_answer):
    global current_question_index, score
    
    # check if the user's answer is correct and update the score accordingly
    if current_answers[int(user_answer)] == correct_answer_list[current_question_index]:
        result_str = "<p>Correct!</p>"
        score += 10
    else:
        result_str = "<p>Wrong.</p>"
        
    # create HTML to display the result and current score
    score_str = f"<p>Score: {score}</p>"
    
    # increment the question index and check if there are more questions
    current_question_index += 1
    
    if current_question_index == len(question_list):
        # if there are no more questions, create HTML to end the game
        end_html = "<p>End of game.</p>"
        html_str = result_str + score_str + end_html
    else:
        # if there are more questions, create HTML to display the next question
        question_html = f"<h2>{question_list[current_question_index]}</h2>"
        answer_html = ""
        current_answers = answer_list[current_question_index]
        for i, answer in enumerate(current_answers):
            answer_html += f"<button onclick='check_answer({i})'>{answer}</button>"
        html_str = result_str + score_str + question_html + answer_html
    
    # display the HTML to the user
    display(HTML(html_str))

# initialize the score and index of the current question
score = 0
current_question_index = 0

# create HTML to display the first question and answers
question_html = f"<h2>{question_list[current_question_index]}</h2>"
answer_html = ""
current_answers = answer_list[current_question_index]
for i, answer in enumerate(current_answers):
    answer_html += f"<button onclick='check_answer({i})'>{answer}</button>"
html_str = question_html + answer_html

# display the HTML to the user
display(HTML(html_str))
```
