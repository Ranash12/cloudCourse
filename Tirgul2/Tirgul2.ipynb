```python
import numpy as np
from google.colab import drive

def read_file_to_array(file_path):
    drive.mount("/content/drive", force_remount=True)
    with open('/content/drive/MyDrive/' + file_path, 'r') as file:
        contents = file.readlines()
    contents = [line.strip() for line in contents]
    return np.array(contents)
```

```python
import ipywidgets as widgets
from IPython.display import display

def create_dropdown(arr):
    dropdown = widgets.Dropdown(options=arr)
    outputs = [widgets.Text() for i in range(4)]
    outputs.append(widgets.Textarea(layout=widgets.Layout(width='50%')))
    update_button = widgets.Button(description='Update')
    
    def dropdown_eventhandler(change):
        selected_values = str(change.new).split(',')
        for i in range(4):
            if i < len(selected_values):
                outputs[i].value = selected_values[i].strip()
            else:
                outputs[i].value = ''
        outputs[-1].value = ''

    def update_button_eventhandler(button):
        with open('/content/drive/MyDrive/Colab Notebooks/CloudCourse/Tirgul2/students.txt', 'r') as f:
            lines = f.readlines()
        with open('/content/drive/MyDrive/Colab Notebooks/CloudCourse/Tirgul2/students.txt', 'w') as f:
            for line in lines:
                if dropdown.value in line:
                    line = f'{outputs[0].description} {outputs[0].value}, {outputs[1].description} {outputs[1].value}, {outputs[2].description} {outputs[2].value}, {outputs[3].description} {outputs[3].value}, {outputs[4].description} {outputs[4].value}\n'
                f.write(line)

    dropdown.observe(dropdown_eventhandler, names='value')
    update_button.on_click(update_button_eventhandler)
    
    display(dropdown)
    display(widgets.VBox(outputs))
    display(update_button)
```

```python
students = read_file_to_array("Colab Notebooks/CloudCourse/Tirgul2/students.txt")

dropdown = create_dropdown(students)
display(dropdown)
```
